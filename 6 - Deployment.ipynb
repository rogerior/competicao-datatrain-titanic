### Utilizar o melhor modelo para realizar a predição dos dados de test

```
%cd drive/My Drive/Estudos/titanic-train
```

```
import pandas as pd
import joblib
import glob
import numpy as np
```

```
dados = pd.read_csv('dados/test_tratado-3_3.csv', sep=';')
dados.head()
```

```
X = dados.drop(columns=['PassengerId', 'Survived'])
X.head()
```

```
caminho = glob.glob('modelos/melhor/best_9573760_1102412_GaussianProcessClassifier*')[0]
caminho
```

```
modelo = joblib.load(caminho)
```

```
predict = modelo.predict_proba(X)
predict
```

```
dados['Survived'] = predict[:, 1]
dados.head()
```
