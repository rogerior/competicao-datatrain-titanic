```
import pandas as pd
```

### Carregar os dados

```
train = pd.read_csv('dados/train.csv')

train.head()
```

```
test = pd.read_csv('dados/test.csv')

test.head()
```

### Juntar arquivos de treino e teste para realizar os tratamentos necessários

```
train['tipo'] = 'train'
test['tipo'] = 'test'

dados = pd.concat([train, test])
dados.head()
```

###Atributo 'Name'
No atributo 'Name', extrair a segunda palavra do nome. Exemplo: 'Hawksford, Mr. Walter James' -> 'Mr.', posteriormente binarizar este atributo

```
def tratarName(name):
  return name.split(',')[1].split('.')[0].strip()
```

```
dados['Name'] = dados['Name'].apply(tratarName)
dados['Name'].unique()
```

### Atributo 'Embarked'
Atribuir valor S para as linhas que contém valor nulo

```
dados['Embarked'].value_counts()
```

```
dados['Embarked'] = dados['Embarked'].fillna('S')
dados['Embarked'].value_counts()
```

### Atributo 'Ticket'
Extrair o primeiro caracter e quando o primeiro caracter for numero atribuir valor N

```
dados['Ticket'] = dados['Ticket'].str[0]
dados['Ticket'].value_counts()
```

```
dados.loc[dados['Ticket'].str.isnumeric() == True, 'Ticket'] = 'N' 
dados['Ticket'].value_counts()
```

### Atributo 'Cabin'
Extrair o primeiro caracter e atribuir valor N para os valores nulos

```
dados['Cabin'] = dados['Cabin'].str[0]
dados['Cabin'].value_counts()
```

```
dados['Cabin'] = dados['Cabin'].fillna('N')
dados['Cabin'].value_counts()
```

### Binarizar atributos
'Pclass', 'Sex', 'Embarked', 'Name', 'Ticket', 'Cabin'

```
dados = pd.get_dummies(dados, columns=['Pclass', 'Sex', 'Embarked', 'Name', 'Ticket', 'Cabin'], drop_first=True)

dados.head()
```

### Atributo 'Age'
Atribuir a média nos valores nulos e remover as casas decimais e criar faixa etária

```
age_mean = dados['Age'].mean()
age_mean
```

```
dados['Age'] = dados['Age'].fillna(age_mean)
```

```
dados['Age'] = dados['Age'].astype('int')

dados.head()
```

```
dados['Age'].unique()
```

```
faixas = [-1, 1, 12, 18, 45, 60, 9999]
labels = [1, 2, 3, 4, 5, 6]
dados['Age'] = pd.cut(dados['Age'], bins=faixas, labels=labels)

dados.head()
```

```
dados['Age'].unique()
```

###Atributo 'Fare'
Criar grupo dos valores

```
dados['Fare'].value_counts().reset_index()#.plot(x='Fare', y='index')
```

```
dados['Fare'].min()
```

```
dados['Fare'].max()
```

```
dados['Fare'].max() / 5
```

```
faixas_fare = [-1, 100, 200, 300, 550]
labels_fare = [1, 2, 3, 4]
dados['Fare'] = pd.cut(dados['Fare'], bins=faixas_fare, labels=labels_fare)

dados.head()
```

```
dados['Fare'].unique()
```

### Atributos 'SibSp' e 'Parch'
Verificar se o passageiro está viajando com algum parente (SibSp + Parch > 0 = 1, else = 0)

```
dados['SibSp'] + dados['Parch'] > 0
```

```
dados['Parente'] = 0
```

```
dados.loc[dados['SibSp'] + dados['Parch'] > 0, 'Parente'] = 1

dados['Parente'].unique()
```

```
dados.head(10)
```

```
dados.drop(columns=['SibSp', 'Parch'], inplace=True)
```

```
dados.dtypes
```

```
dados['Age'] = dados['Age'].astype('int')
dados['Fare'] = dados['Fare'].astype('int')
```

### Separar os dados de treino e teste

```
tratado_train = dados[dados.tipo == 'train'].copy()
tratado_train.drop(columns=['tipo'], inplace=True)
tratado_train.head()
```

```
tratado_train.shape
```

```
tratado_test = dados[dados.tipo == 'test'].copy()
tratado_test.drop(columns=['tipo'], inplace=True)
tratado_test.head()
```

```
tratado_test.shape
```

### Verificar correlação dos atributos 
para uma possível seleção de apenas os atributos mais relevantes

```
corr = tratado_train.corr()
corr
```

```
corr['Survived'].sort_values()
```

```
# Atributos com maior correlação
selecao = corr[(corr['Survived'] > 0.15) | (corr['Survived'] < -0.15)]['Survived']
selecao
```

```
atributos_selecionados = list(selecao.index)
atributos_selecionados.append('PassengerId')
atributos_selecionados
```

```
tratado_train = tratado_train[atributos_selecionados]

tratado_train.head()
```

```
tratado_test = tratado_test[atributos_selecionados]

tratado_test.head()
```
